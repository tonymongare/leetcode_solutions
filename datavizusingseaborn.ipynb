```python
#importign all the necessary libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#line plot
tips = sns.load_dataset('tips')
print(tips.head)
print(tips.tail)

```

```python
tips.columns
```

```python
#plotting the line plot
sns.lineplot(data=tips, y='tip', x='day', hue='smoker')
plt.title = "Tips throught the week for both smokers and non smokers"
plt.show()
```

```python
tips.columns
```

```python
sns.relplot(data = tips, x='day', y='tip', hue='sex',kind='line')
```

```python
#box plot
tips.columns
```

```python
sns.barplot(data=tips, x='total_bill', y='sex', hue='day', orient = 'h');

```

```python
#you can alos use this method
tips.columns
```

```python
sns.lineplot(data=tips, x="sex", y="total_bill", hue="sex")
sns.barplot(data=tips, x="sex", y="total_bill", hue="sex")
```

```python
#cat plot  is use to compare relationship between numerical and one or more cateogical variables

sns.catplot(data=tips, x="day", y="total_bill")
```
